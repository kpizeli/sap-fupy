```python
from string import Template
import pandas as pd
```

```python
df = pd.read_excel('relatorio_aprovadores.xlsx', parse_dates=['Vencimento'])
```

```python
df = df.loc[(df['Status Tarefa'] == 'Ag. Aprovaﾃｧﾃ｣o')&(df['Tipo Documento'] == 'Contrato de Compras')].reset_index(drop=True)
```

```python
df.sort_values('Documento')
```

```python
# O seu layout prﾃｩ-determinado (a "t-string")
LAYOUT_MENSAGEM = Template("""
東 *SOLICITAﾃﾃグ DE APROVAﾃﾃグ*
---------------------------------------
塘 Documento: $doc
召 Fornecedor: $fornecedor
腸 Centro de Custo: $centro_custo
套 Vencimento: $vencimento
---------------------------------------
Favor verificar a pendﾃｪncia no SAP.
""")
```

```python
for index, row in df.iterrows():

    dados_do_fup = {
    'doc': str(row['Documento']).split('.')[0],
    'fornecedor': row['Nome do fornecedor'],
    'centro_custo': row['Centro de Custo'],
    'vencimento': row['Vencimento'] # Aqui vocﾃｪ pode formatar a data se desejar
}
    mensagem_final = LAYOUT_MENSAGEM.substitute(dados_do_fup)
    
    print(mensagem_final)
```
